```python
# Import Dependencies
import pandas as pd 
import numpy
import requests
import json
import datetime as dt
```

# From open data to retrieve lat and lon of station_id 

```python
url = "https://tor.publicbikesystem.net/ube/gbfs/v1/en/station_information"
print(requests.get(url))
```

```python
# Pretty Print the output of the JSON
response = requests.get(url).json()
#print(json.dumps(response, indent=4, sort_keys=True))
station_id = response['data']['stations'][0]['station_id']
lat = response['data']['stations'][0]['lat']
lon = response['data']['stations'][0]['lon']
```

```python
# Retrieve data: station_id, lat, lon 
stations = response['data']['stations']
lat = response['data']['stations']
lon = response['data']['stations']

station_id = []
station_name =[]
lats = []
lons = []
for station in stations: 
    #print(station["station_id"])
    station_id.append(station["station_id"])
    station_name.append(station["name"])
    lats.append(station['lat'])
    lons.append(station['lon'])
```

```python
# Create data frame 
station_coordinate = pd.DataFrame({"station_id":station_id,"lat":lats,"lon":lons})
station_coordinate
```

```python
# Create data frame 
stations = pd.DataFrame({"station_id":station_id,"station_name":station_name})
#stations = stations.set_index(['station_id'])
stations
```

# Import and clean the quarterly bikeshare files

```python
q1_file_path = "../Project 2/2017 Data/Bikeshare Ridership (2017 Q1).csv"
q1 = pd.read_csv(q1_file_path)
q1.head()
```

```python
q2_file_path = "../Project 2/2017 Data/Bikeshare Ridership (2017 Q2).csv"
q2 = pd.read_csv(q2_file_path)
q2.head()
```

```python
# Merge q1 & q2 
q1q2 = pd.concat([q1,q2])

# Change to datetime 
q1q2['trip_start_time'] = pd.to_datetime(q1q2['trip_start_time'],format = '%d/%m/%Y %H:%M')
q1q2['trip_stop_time'] = pd.to_datetime(q1q2['trip_stop_time'],format = '%d/%m/%Y %H:%M')
q1q2.head()
```

```python
# Check if any missing values in q1q2
q1q2.isnull().sum()
```

```python
q3_file_path = "../Project 2/2017 Data/Bikeshare Ridership (2017 Q3).csv"
q3 = pd.read_csv(q3_file_path)

#Change trip_start_time and trip_stop_time to datetime 
q3['trip_start_time'] = pd.to_datetime(q3['trip_start_time'],format = '%m/%d/%Y %H:%M')
q3['trip_stop_time'] = pd.to_datetime(q3['trip_stop_time'],format = '%m/%d/%Y %H:%M')
q3.head()
```

```python
q4_file_path = "../Project 2/2017 Data/Bikeshare Ridership (2017 Q4).csv"
q4 = pd.read_csv(q4_file_path)
q4 = q4.dropna()

#Change trip_start_time and trip_stop_time to datetime format 
q4['trip_start_time'] = pd.to_datetime(q4['trip_start_time'],format = '%m/%d/%y %H:%M:%S')
q4['trip_stop_time'] = pd.to_datetime(q4['trip_stop_time'],format = '%m/%d/%y %H:%M:%S')
q4.head()
```

```python
# Merge q3 & q4
q3q4 = pd.concat([q3,q4])
q3q4.head()
```

```python
# Check if any missing values in q3q4
q3q4.isnull().sum()
```

```python
add_station_id_to_q3q4 = q3q4.merge(stations, left_on = "from_station_name", right_on ="station_name", how = "left")\
   .merge(stations, left_on = "to_station_name", right_on ="station_name",how = "left")

add_station_id_to_q3q4.head()
```

```python
q3q4 = add_station_id_to_q3q4

#Rename columns name
q3q4 = q3q4.rename(columns={"station_id_x": "from_station_id","station_id_y": "to_station_id"})

#Reposition columns
q3q4 = q3q4[['trip_id','trip_start_time', 'trip_stop_time','trip_duration_seconds','from_station_id', 'from_station_name','to_station_id', 'to_station_name', 'user_type']]
q3q4
```

```python
q3q4 = q3q4.dropna()
q3q4.isnull().sum()
```

```python
# Merge quarters to a year 
bikes_data_2017 =  pd.concat([q1q2 ,q3q4])
bikes_data_2017
```

```python
bikes_data_2017.isnull().sum()
```

```python
# Combine stations coordinate to bikes_data_2017
final_bikes_data_2017 = bikes_data_2017.merge(station_coordinate, left_on = "from_station_id", right_on ="station_id", how = "left")\
   .merge(stations, left_on = "to_station_id", right_on ="station_id",how = "left")
final_bikes_data_2017.dropna()
```
